```python
X=fraud[['TransactionAmt','card1','addr1','addr2',
        'C1','C2','C3','C4','C5','C6','C7','C8','C9','C10','C11','C12','C13','C14',
        'V95', 'V96', 'V97','V98', 'V99', 'V100', 'V101', 'V102', 'V103', 'V104', 'V105', 'V106', 'V107', 'V108', 'V109', 'V110', 'V111',
        'V112', 'V113', 'V114', 'V115', 'V116', 'V117', 'V118', 'V119', 'V120', 'V121', 'V122', 'V123', 'V124', 'V125', 'V126' , 'V127',
        'V128', 'V129', 'V130', 'V131', 'V132', 'V133', 'V134', 'V135', 'V136', 'V137']]

y=fraud['isFraud']
```

### Under-sampling data balancing  <a class="anchor" id="fUndataBalancing"></a>

[TOC](#ftop)

The imbalanced datasets are a characteristic of the problem of credit card fraud detection. So, we must use a technique to balance the data. I used the two possible methods: under-sampling and upper-sampling. After the under-sampling technique, the remaining data has over 10 thousand legitimate and fraudulent transactions.

```python
#Let's perform an undersampling to balance the data. In this case, it was used a technique of random sampling with replacement.
rus = RandomUnderSampler(random_state=42, replacement=True)# fit predictor and target variable
x_rus, y_rus = rus.fit_resample(X, y)
```

```python
#Let's visually see the balanced data in a simple bar chart.
bar_chart = alt.Chart(y_rus.value_counts().reset_index().rename(columns={0:'counts'})).mark_bar().encode(
    y=alt.Y('isFraud:Q', title='Count'),
    x=alt.X('index:N', axis=alt.Axis(labelAngle=0), title='isFraud'))
bar_chart
text_chart = bar_chart.mark_text(align='left', baseline='bottom', dx=-5).encode(text='isFraud:Q')
(bar_chart + text_chart).properties(height=300, width=400, title='Balanced remaning data')
```

Now, the two classes (legitimate and fraudulent) are balanced.

I designed the following graphic to check the amount purchased distribution of the remaining data. It is possible to verify the Poisson distribution. It's essential and intuitive information that cheap products usually sell many more units than expensive products.  

```python
spd = x_rus[['TransactionAmt']].sample(n = 500)
spd.hist(figsize=[6, 4])
plt.title('Purchase amount distribution')
plt.xlabel("Purchased amount")
plt.ylabel("Count")
plt.show()
```

I ran a hypothesis test for Poisson distribution to check if the average purchased in fraudulent transactions is higher than in legitimate transactions. 
The null and alternative hypotheses are designed as follows:

$H_0:\mu _{fraud} \le \mu _{legitimate} $ (The average purchase amount of fraudulent transactions is not higher than legitimate transactions).

$H_1:\mu _{fraud} > \mu _{legitimate} $ (The average purchase amount of fraudulent transactions is higher than legitimate transactions).

```python
legitimateTr=x_rus[y_rus == 0]['TransactionAmt'].sample(n = 500)
fraudTr=x_rus[y_rus == 1]['TransactionAmt'].sample(n = 500)
```

Applying the poisson mean test to check, based on a p-value under 5%, one can conclude that we have to reject the null hypothesis, i.e., the average purchased amount of fraudulent transactions is higher than legitimate transactions.

```python
stats.poisson_means_test(len(fraudTr), sum(fraudTr), len(legitimateTr), sum(legitimateTr),
                        diff=0.0, alternative='less')
```

### Feature Selection  <a class="anchor" id="fUnfeatureSelection"></a>

[TOC](#ftop)

Once the data is balanced, one can perform feature selection. For this task, one chose the Recursive feature elimination with cross-validation to select features. This technique performs recursive feature elimination in a cross-validation loop to find the optimal number of features. Given an external estimator that assigns weights to features, recursive feature elimination aims to select features by recursively considering smaller and smaller sets of features. First, it trains the estimator on the initial set of features, and each feature's importance is obtained either through any specific attribute or callable. Then, it prunes the most minor essential features from the current features. That procedure is recursively repeated on the pruned set until it reaches the desired number of features.

```python
X_train, X_test, y_train, y_test = train_test_split(x_rus, y_rus,
                                                    test_size=0.2,
                                                    random_state=42)
```

```python
min_features_to_select = 1  # Minimum number of features to consider
clf = RandomForestClassifier(n_estimators=10,
                             random_state=42)

cv = StratifiedKFold(5)

rfecv = RFECV(
    estimator=clf,
    step=1,
    cv=cv,
    scoring="accuracy",
    min_features_to_select=min_features_to_select,
    n_jobs=2,
)
rfecv.fit(X_train, y_train)

print(f"Optimal number of features: {rfecv.n_features_}")
print(f"Selected features: {X_train.columns[rfecv.support_]}")
```

One could verify the 13 selected fields. Those fields will be used as input for the under-sampling machine learning models in the modeling step of the project. Please check the project's modeling notebook. 
The following graphic plots the gain by recursive field elimination with cross-validation. One can see that there is gain just until 13 selected fields.

```python
n_scores = len(rfecv.cv_results_["mean_test_score"])
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Mean test accuracy")
plt.errorbar(
    range(min_features_to_select, n_scores + min_features_to_select),
    rfecv.cv_results_["mean_test_score"],
    yerr=rfecv.cv_results_["std_test_score"],
)
plt.title("Recursive Fraud Feature Elimination \nwith correlated features")
plt.show()
```

### Upper-sampling data balancing  <a class="anchor" id="fUpdataBalancing"></a>

[TOC](#ftop)

The second technique used upper-sampling the balance the dataset. This technique used SMOTE that stands for Synthetic Minority Oversampling Technique. After the upper-sampling the amount of fraudulent and legitimate transactions are the same, 420887 transactions each.

```python
X_resampled, y_resampled = SMOTE().fit_resample(X, y)
```

The following graphic illustrates the amount of dataset transactions. The balanced dataset is ready to be used by a machine-learning model. 

```python
transGroup1=pd.concat([X_resampled, y_resampled], axis=1)[['TransactionAmt','isFraud']].groupby(
    'isFraud').agg('count').reset_index()
transGroup1.columns = ['isFraud','count']
```

```python
bar_chart=alt.Chart(transGroup1).mark_bar().encode(
    alt.X('isFraud:N', axis=alt.Axis(labelAngle=0)),
    alt.Y('count:Q', title='Count'))
text_chart = bar_chart.mark_text(align='left', baseline='bottom', dx=-5).encode(text='count:Q')
(bar_chart + text_chart).properties(height=300, width=400, title='Total count of re-sampled transactions')
```

I designed the following graphic to check the amount purchased distribution of the generated data. It is possible to verify that the data maintains the Poisson distribution. It's essential and intuitive information that cheap products usually sell many more units than expensive products, so SMOTE has to keep this characteristic. 

```python
spd = X_resampled[['TransactionAmt']].sample(n = 500)
spd.hist(figsize=[6, 4])
plt.title('Purchase amount distribution')
plt.title('Purchase amount distribution')
plt.xlabel("Purchased amount")
plt.ylabel("Count")
plt.show()
```

I also ran a hypothesis test for Poisson distribution to check if the synthetic average purchased in fraudulent transactions is still higher than in legitimate transactions. 
The null and alternative hypotheses are designed as follows:

$H_0:\mu _{fraud} \le \mu _{legitimate} $ (The average purchase amount of fraudulent transactions is not higher than legitimate transactions).

$H_1:\mu _{fraud} > \mu _{legitimate} $ (The average purchase amount of fraudulent transactions is higher than legitimate transactions).

```python
legitimateTr=X_resampled[y_resampled == 0]['TransactionAmt'].sample(n = 500)
fraudTr=X_resampled[y_resampled == 1]['TransactionAmt'].sample(n = 500)
```

Applying the Poisson mean test again, based on a p-value under 5%, one can conclude that we have to reject the null hypothesis, i.e., the average purchase of fraudulent transactions is higher than legitimate transactions. Verifying that SMOTE keeps the same behavior as the original data is imperative.

```python
stats.poisson_means_test(len(fraudTr), sum(fraudTr), len(legitimateTr), sum(legitimateTr),
                        diff=0.0, alternative='less')
```

### Feature Selection  <a class="anchor" id="fUpfeatureSelection"></a>

[TOC](#ftop)

Once the data is balanced, one can perform feature selection for the upper-sampling case. One keeps the same Recursive feature elimination with cross-validation to select features. As already presented, this technique performs recursive feature elimination in a cross-validation loop to find the optimal number of features. Given an external estimator that assigns weights to features, recursive feature elimination aims to select features by recursively considering smaller and smaller sets of features. First, it trains the estimator on the initial set of features, and each feature's importance is obtained either through any specific attribute or callable. Then, it prunes the most minor essential features from the current features. That procedure is recursively repeated on the pruned set until it reaches the desired number of features.

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled,
                                                    test_size=0.2,
                                                    random_state=42)
```

```python
min_features_to_select = 1  # Minimum number of features to consider
clf = RandomForestClassifier(n_estimators=10,
                             random_state=42)

cv = StratifiedKFold(5)

rfecv = RFECV(
    estimator=clf,
    step=1,
    cv=cv,
    scoring="accuracy",
    min_features_to_select=min_features_to_select,
    n_jobs=2,
)
rfecv.fit(X_train, y_train)

print(f"Optimal number of features: {rfecv.n_features_}")
print(f"Selected features: {X_train.columns[rfecv.support_]}")
```

One could verify the 14 selected fields. Those fields will be used as input for the upper-sampling machine learning models in the modeling step of the project. Please check the project's modeling notebook. 
The following graphic plots the gain by recursive field elimination with cross-validation. One can see that there is gain just until 14 selected fields.

```python
n_scores = len(rfecv.cv_results_["mean_test_score"])
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Mean test accuracy")
plt.errorbar(
    range(min_features_to_select, n_scores + min_features_to_select),
    rfecv.cv_results_["mean_test_score"],
    yerr=rfecv.cv_results_["std_test_score"],
)
plt.title("Recursive Fraud Feature Elimination \nwith correlated features")
plt.show()
```
